```python
employee = pd.read_csv("data/employees.csv", parse_dates = ['Start Date', 'Last Login Time'])
employee['Senior Management'] = employee['Senior Management'].astype('bool')
employee['Gender'] = employee['Gender'].astype('category')
employee['Team'] = employee['Team'].astype('category')
# employee.info()
employee.head()
```

```python
# Filter a DataFrame Based on A Condition
employee[employee['Gender']  == "Male"]
```

```python
employee[employee['Team'] == "Finance"]
```

```python
mask = employee['Team'] == "Finance"
employee[mask]
```

```python
employee[employee['Senior Management']]
```

```python
mask = employee['Team'] != 'Marketing'
employee[mask]
```

```python
mask = employee['Salary'] >= 100000
employee[mask]
```

```python
# Filter DataFrame with More than One Condition (AND - &)
mask1 = employee['Salary'] >= 100000
mask2 = employee['Gender'] == 'Male'
mask3 = employee['Team'] != 'Legal'
employee[mask1 & mask2 & mask3]
```

```python
# Filter DataFrame with More than One Condition (OR - |)
mask1 = employee['Start Date'] < '1990-01-01'
mask2 = employee['Senior Management']
employee[mask1 | mask2]
```

```python
# Check for Inclusion with the isin Method
mask1 = employee['Team'] == 'Marketing'
mask2 = employee['Team'] == 'Finance'
mask3 = employee['Team'] == 'Engineering'
employee[mask1 | mask2 | mask3]
```

```python
employee = pd.read_csv("data/employees.csv")
mask = employee['Team'].isin(['Marketing', 'Finance', 'Sales'])
employee[mask]
```

```python
# Check for Null and Present DataFrame Values with the isnull and notnull Methods
employee = pd.read_csv("data/employees.csv", parse_dates = ['Start Date', 'Last Login Time'])
employee['Senior Management'] = employee['Senior Management'].astype('bool')
employee['Gender'] = employee['Gender'].astype('category')
employee['Team'] = employee['Team'].astype('category')
# employee.info()
employee.head()
```

```python
employee['Team'].isnull()
```

```python
mask = employee['Team'].isnull()
employee[mask]
```

```python
mask = employee['Team'].notnull()
employee[mask]
```

```python
# Check For Inclusion Within a Range of Values with the between Method
mask = employee['Salary'].between(60000, 70000)
employee[mask]
```

```python
mask = employee['Bonus %'].between(2.0, 5.0)
employee[mask]
```

```python
mask = employee['Start Date'].between('1990-01-01', '1992-01-01')
employee[mask]
```

```python
mask = employee['Last Login Time'].between('08:30AM', '04:30PM')
employee[mask]
```
