## Interviewprep data analysis using Python - Merge, Join, Concat

```python
import pandas as pd
import numpy as np
```

```python
week1 = pd.read_csv("data/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("data/Restaurant - Week 2 Sales.csv")
customer = pd.read_csv("data/Restaurant - Customers.csv")
foods = pd.read_csv("data/Restaurant - Foods.csv")
# foods
```

```python
# The pd.concat Method, Part 1
# pd.concat(objs = [week1, week2])
pd.concat(objs = [week1, week2], ignore_index = True)
```

```python
week1.append(other = week2, ignore_index = True)
```

```python
sales = pd.concat(objs = [week1, week2], keys = ['week 1', 'week 2'])
sales
```

```python
sales.loc[('week 1',2)]
```

```python
sales.loc[('week 1',2), 'Food ID']
```

```python
# Inner Joins, Part 1
week1.merge(week2, how='inner', on = 'Customer ID')
```

```python
week1.merge(week2, how='outer', on = 'Customer ID')
```

```python
week1.merge(week2, how='left', on = 'Customer ID')
```

```python
week1.merge(week2, how='right', on = 'Customer ID')
```

```python
week1.merge(week2, how='inner', on = 'Customer ID', suffixes = ['- A', '- B'])
```

```python
# The left_on and right_on Parameters
```

```python
pd.merge(week1, customer, how = 'left', left_on = 'Customer ID', right_on = 'ID')
```
