```python
# A Review of the apply Method on a pandas Series Object
def revenue_status(row):
    
    revenue = row['Box Office']
    
    if revenue >= 800:
        return "High Revenue"
    elif 500 < revenue < 800:
        return "Medium Revenue"
    else:
        return "Low Revenue"

bond.apply(revenue_status, axis = 1)
```
